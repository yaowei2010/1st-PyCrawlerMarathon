### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
# Total Classification
for d in soup.find(class_="part_menu_2").find_all('a'):
    print(d.text)
    
```

```python
#Show Date, Classification, and Title of the News
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text,d.find('em').text, d.find_all('a')[-1].text)
```

```python
import pandas as pd
news = list()
for d in soup.find(class_="part_list_2").find_all('h3'):
    news_time = d.find(class_="date").text
    news_type = d.find('em').text
    news_title = d.find_all('a')[-1].text
    record = [news_time,news_type,news_title]
    news.append(record)

news_data = pd.DataFrame(news)
news_data.columns = ["news_time","news_type","news_title"]
```

```python
print(news_data)
```

### ③ 哪一個類別的文章最多

```python
import matplotlib.pyplot as plt
x = news_data['news_type'].groupby(news_data['news_type']).count()
#print("The most abudance type of news today is ", max(x))
print(x)
```
