## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
from selenium.webdriver.chrome.options import Options
from datetime import datetime, timedelta
import pandas as pd

#all date
date_time = datetime.now() - timedelta(1)
date_time = date_time.strftime("%Y/%m/%d")
print("取得昨天時間：", date_time)


options = Options()
options.add_argument('--disable-notifications')

browser = webdriver.Chrome(executable_path='./chromedriver',options = options)

browser.get("https://www.ettoday.net/news/news-list.htm")


while True:
    browser.execute_script("window.scrollTo(0, 100000);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    print(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text)
    if date_time in soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text:
        print('==== STOP ====')
        break    
r = browser.page_source
    
soup = BeautifulSoup(r, "html5lib")

record = []


for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
    news = {'time':d.find(class_="date").text, 'title':d.find_all('a')[-1].text}
    record.append(news)

df = pd.DataFrame(record)

browser.close()
```

## 2. 取出現在時間兩小時內的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
from selenium.webdriver.chrome.options import Options
from datetime import datetime, timedelta
import pandas as pd
browser = webdriver.Chrome(executable_path='./chromedriver',options = options)

browser.get("https://www.ettoday.net/news/news-list.htm")
# Within two hours

target_time = datetime.now() - timedelta(hours=2)


while True:
    browser.execute_script("window.scrollTo(0, 100000);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    print(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text)
    local_time = datetime.strptime(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text,'%Y/%m/%d %H:%M')
    if local_time < target_time :
        print('==== STOP ====')
        break    
r = browser.page_source
    
soup = BeautifulSoup(r, "html5lib")

record = []


for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
    news = {'time':d.find(class_="date").text, 'title':d.find_all('a')[-1].text}
    record.append(news)

df2 = pd.DataFrame(record)

browser.close()
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
from selenium.webdriver.chrome.options import Options
from datetime import datetime, timedelta
import pandas as pd


date_time = datetime.now() - timedelta(3)
## the webpage cannot select by number with 0 after the true number, e.g: 06, 02
date_M = str(int(date_time.strftime("%m")))
date_D = str(int(date_time.strftime("%d")))
timestart = datetime.strptime(date_time.strftime("%Y/%m/%d") + ' 15', '%Y/%m/%d %H')
timeend = datetime.strptime(date_time.strftime("%Y/%m/%d") + ' 17', '%Y/%m/%d %H')

browser = webdriver.Chrome(executable_path='./chromedriver',options = options)
browser.get("https://www.ettoday.net/news/news-list.htm")


# 選擇且點選日期
from selenium.webdriver.support.ui import Select

selM = Select(browser.find_element_by_id("selM"))
selM.select_by_value(date_M)
selD = Select(browser.find_element_by_id("selD"))
selD.select_by_value(date_D)
browser.find_element_by_id('button').click()


while True:
    browser.execute_script("window.scrollTo(0, 100000);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    print(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text)
    local_time = datetime.strptime(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text,'%Y/%m/%d %H:%M')
    if local_time < timestart :
        print('==== STOP ====')
        break    
r = browser.page_source
    
soup = BeautifulSoup(r, "html5lib")

record = []


for d in soup.find(class_="part_list_2").find_all('h3'):
    local = datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M')
    if local > timestart and local < timeend:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
        news = {'time':d.find(class_="date").text, 'title':d.find_all('a')[-1].text}
        record.append(news)

df3 = pd.DataFrame(record)




browser.close()
```
