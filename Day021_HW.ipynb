### 2. 剛剛的做法只能取出第一頁，試著取出所有頁面的資料。

```python
# 1. 先取出全部有幾頁
import re
links = []
for d in soup.find_all("li",class_="rc-pagination-item"):
    link = "https://hahow.in"+d.a["href"]
    links.append(link)
    
#看最後一個
match = re.findall('(\d+)',links[len(links)-1])
allpage=int(match[0])
print(allpage)
```

```python
# 2. 每頁都取出資料放入 courses
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
browser = webdriver.Chrome(executable_path='chromedriver')

courses = []

for i in range(1,allpage):
    url = "https://hahow.in/courses?page=" + str(i)
    print(url)
    browser.get(url)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'html.parser')

    for d in soup.find_all("h4"):
        print(d.text)
        courses.append(d.text)

browser.close()
```

```python
# 3. done!

print(courses)
```
