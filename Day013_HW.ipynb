### ① 印出最新文章的「作者」「標題」「時間」

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/NBA/index.html'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

import pandas as pd

def GetUrlArtical(url):
    r = requests.get(url)
    column_name = ['Author','Channel','Title','Time']
    soup = BeautifulSoup(r.text, "html5lib")
    articles = []
    for d in soup.find_all(class_="title"):
    #print(d.text.replace('\t', '').replace('\n', ''))
        try:
            r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")
            record = []
            for info in r.find_all(class_='article-meta-value'):
                record.append(info.text)
            articles.append(record)
        except:
            continue
    df = pd.DataFrame(articles, columns=column_name)
    return df
```

```python
url1 = 'https://www.ptt.cc/bbs/NBA/index.html'
GetUrlArtical(url1)
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」
