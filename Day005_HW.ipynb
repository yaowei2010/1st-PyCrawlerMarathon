```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests, json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

import pandas as pd
data = pd.read_json(response)
data.head()

```

```python
print(len(data))
print(data.columns.values)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

data2 = data[['title', 'createdAt', 'commentCount', 'likeCount']]

data2
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」


r_npop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
n_response = r_npop.text

import pandas as pd
n_data = pd.read_json(n_response)

print('熱門文章-平均留言人數:', data['commentCount'].mean())
print('熱門文章-平均按讚人數:', data['likeCount'].mean())
print('非熱門文章-平均留言人數:', n_data['commentCount'].mean())
print('非熱門文章-平均按讚人數:', n_data['likeCount'].mean())
```
